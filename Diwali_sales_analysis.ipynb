# Exploratory Data Analysis

## Gender Analysis

```python
df.columns
```

```python
# plotting a bar chart for gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender  vs amount.

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

*From above graphs we can see that most of the buyers are females and earn the purchasing power of females are greater than male*

## Age Analysis

```python
# plotting a bar chart for age group and it's count with legends gender.

ax = sns.countplot(data = df, x = 'Age Group', hue= 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for age group vs amount

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ax = sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)

for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female* 

## State Analysis

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc = {'figure.figsize': (16,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
# total amount/ sales from top 10 states 
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc = {'figure.figsize': (16,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
```

*From above graphs we can see that most of the orders & total sales / amount are from Uttar Pradesh, Maharashtra, and Karnataka respectively*

## Marital Status

```python
# plotting a bar chart for marital_status and it's count

ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc = {'figure.figsize': (7,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for sales analysis and grouped by marital status & gender wise

sales_Marital_Status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc = {'figure.figsize': (7,4)})
sns.barplot(data = sales_Marital_Status, x = 'Marital_Status', y  = 'Amount', hue = 'Gender')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power.*

## Occupation Analysis

```python
# plotting a bar chart for Occupation and it's count

ax = sns.countplot(data = df, x = 'Occupation')

sns.set(rc = {'figure.figsize':(30, 5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for sales with grouped by occupation

sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.set(rc = {'figure.figsize': (20,5)})
sns.barplot(data = sales_occupation, x = 'Occupation', y = 'Amount')
```

*From above graphs we can see that most of the buyers are working IT, Health Care & Aviation sector*

## Product Category Analysis

```python
# plotting a bar chart for product categorry and it's count

ax = sns.countplot(data = df, x = 'Product_Category')

sns.set(rc = {'figure.figsize': (30, 4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for top 10 sales analysis with group by product categories.

sales_product_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)

sns.set(rc = {'figure.figsize': (20,5)})
sns.barplot(data = sales_product_category, x = 'Product_Category', y = 'Amount')
```

*From above graphs we can see that most of the sold products are from Food ,clothing, and electonics categories.*

## Product id analysis

```python
# plotting a bar chart for order wise with group by product id 

sales_productid = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc = {'figure.figsize': (20, 6)})
sns.barplot(data = sales_productid, x = 'Product_ID', y = 'Orders')
```

```python
# Top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12, 7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
# Multiple Aggregation

state_analysis = df.groupby('State')['Amount'].agg(
    Total_Sales = 'sum',
    Average_Sales = 'mean',
    Max_Sales = 'max',
    Transactions = 'count'
).sort_values('Total_Sales', ascending=False)

state_analysis.head()
```

## Insights
- Uttar Pradesh generated the highest total revenue
- Provides sales performance matrics for each state

```python
# Using pivot table
pivot_gender_state = pd.pivot_table(
    df,
    values = 'Amount',
    index = 'State',
    columns = 'Gender',
    aggfunc = 'sum',
    fill_value = 0
)

pivot_gender_state.head()
```

## Insigths
- Female spending exceeds male spending in most states
- Helps identified gender specific purchasing trends

```python
# Who are our top customers by spending.

top_customers = (
    df.groupby('User_ID', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
)
top_customers
```

```python
# Customer Ranking

top_customers['Rank'] = top_customers['Amount'].rank(
    ascending = False,
    method = 'dense'
)
top_customers.head(10)
```

## Insights
- Ranks customers based on total spending and helps identify high value customers

```python
## Occupation wise Average purchases

occupation_avg =(
    df.groupby('Occupation')['Amount'].mean().sort_values(ascending=False).head(10)
)
occupation_avg
```

## Insights

```python
# Create Spending Category using Apply

def category(amount):
    if amount < 5000:
        return 'Low'
    elif amount < 15000:
        return 'Medium'
    else :
        return 'High'

df['Spending_Category'] = df['Amount'].apply(category)

df[['Amount', 'Spending_Category']].head(10)
```
